```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
```

```python
pca.fit(dff)
```

```python
print(pca.components_)
```

```python
print(pca.explained_variance_)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

pipeline = Pipeline([('scaling', StandardScaler()), ('pca', PCA(n_components=3))])
```

```python
PC = pipeline.fit_transform(dff)
```

```python
PC
```

```python
from mpl_toolkits import mplot3d
```

```python
ax = plt.axes(projection='3d')
ax.scatter3D(PC[:,0],PC[:,1],PC[:,2])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
```

```python
x = dff.values
x
```

```python
from sklearn.preprocessing import scale
x = scale(x);x
```

```python
pca.fit(x)
```

```python
var = np.array(pca.explained_variance_)
var
```

```python
exp = np.sum(var)/11
```

```python
exp= exp*100
```

```python
pca.components_
```

```python
y = pca.transform(x)
y
```

```python
from sklearn.cluster import KMeans
ks = range(1, 10)
inertias = []
for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(y)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

Elbow is formed for k = 4

```python
import scipy.cluster.hierarchy as shc
plt.title("Cars Dendograms")
dend = shc.dendrogram(shc.linkage(x, method='ward'))
```

Using this dendogram, we get the number of clusters to be equal to 4 and we cut the dendogram at resolution = 7.5

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
label2=cluster.fit_predict(x)
```
